```python
import numpy as np 
import pandas as pd 
```

## Find missing data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data) #np.nan means null or not a number value
```

```python
df
```

```python
df.isna() #find where value is null
```

```python
df.isna().sum() #sum of null values in a col
```

```python
df.isna().any() #find cols if they have any null avlues
```

## Removing Missing Data

```python
df
```

```python
df.dropna() #removes row having null values
```

```python
df.dropna(thresh=3) #atleast 3 non null values
```

## Filling the missing data

```python
df.fillna(0) #fill null values with 0
```

```python
values = {'A':0, 'B' : 100, 'C' : 300, 'D' : 400} #specify what data we want to fill
df.fillna(values) 
```

```python
df.fillna(df.mean()) #fill values with mean values of na
```
