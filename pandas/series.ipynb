## Series in pandas: 
A series is a 1d array capable of storing any data type. This axis labels are collectively knows as index. These are stacked vertically. Ways of creating series:

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
my_list = [10, 20, 30]
arr = np.array(my_list)
d = {1:10, 2:20, 3:30}
```

```python
pd.Series(my_list) #labels : data points
```

```python
pd.Series(my_list, index=labels) #using custom labels
```

```python
pd.Series(arr) #series is 1d, 2d gives error
```

```python
pd.Series(d) #key : value_pair type dictionary
```

We have to deal with multiple series, group of series = data frames
