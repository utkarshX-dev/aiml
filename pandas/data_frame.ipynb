## Data frames: Group of series

```python
#Creating a data frame
```

```python
import numpy as numpy
import pandas as pd
```

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
} #dictionary of key-value pairs
df = pd.DataFrame(data) #group of series
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
df2
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
columns = ["Name","Age","City","Salary"]
df2 = pd.DataFrame(data_list,columns =columns) #using custom column name
df2
```

```python
#Selection and indexing of columns
```

```python
df2["City"]
```

```python
df2[["Name", "City"]] #accessing multiple row column at a time
```

```python
#Creating a new column
```

```python
df2["Designation"] = ["Doctor","Eng.","Doctor","Eng."]
```

```python
df2
```

```python
#removing column
```

```python
df2.drop("Designation", axis= 1) #this doesn't works in inplace i.e this creates a copy
```

```python
df2
```

```python
df2.drop("Designation", axis=1, inplace=True) #in-place column removal
df2
```

```python
df2.drop(0, axis= 0) #row can be delete by using axis = 0
```
